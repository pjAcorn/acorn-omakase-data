```python
# URL
SeoulURL = "http://api.odcloud.kr/api/15083033/v1/uddi:324125cb-6185-41a8-9480-8be3a8e4a717"
params = {'serviceKey': decodingKey, 'pageNo': 1, 'numOfRows': 10, 'perPage': 334}

SeoulResponse = requests.get(SeoulURL, params=params)
SeoulData = SeoulResponse.text

SeoulJsonData = json.loads(SeoulData)
SeoulDf = pd.json_normalize(SeoulJsonData['data'])

# 유아교육과 학원 분류 생성
SeoulDf['분류'] = SeoulDf.apply(lambda x: '학원' if x['업종중분류']!='유아교육' else '유아교육', axis=1)
# 학원 분류일 때 앞에 붙는 '학원-' 삭제
SeoulDf['업종중분류'] = SeoulDf.apply(lambda x: x['업종중분류'][3:] 
                                     if x['분류']=='학원' else x['업종중분류'], axis=1)
SeoulDf['업종소분류'] = SeoulDf.apply(lambda x: x['업종소분류'][3:] 
                                     if x['분류']=='학원' else x['업종소분류'], axis=1)

# 컬럼명/순서 변경
SeoulDf.rename({'광역시도':'시도'}, axis=1, inplace=True)
SeoulDf = SeoulDf.reindex(['시도', '시군구', '업종대분류', '분류', '업종중분류', '업종소분류', '건수'], axis=1)

# CSV 저장
SeoulDf.to_csv("Seoul.csv", index=False, encoding='utf-8-sig')
SeoulDf
```

* 상권별 업종 밀집 통계 2017

```python
# URL
yearsURL = "http://api.odcloud.kr/api/15083033/v1/uddi:ad52aabc-3723-428c-874b-90db571eeb6e"
params = {'serviceKey': decodingKey, 'pageNo': 1, 'numOfRows': 10, 'perPage': 20625}

yearsResponse = requests.get(yearsURL, params=params)
yearsData = yearsResponse.text

yearsJsonData = json.loads(yearsData)
yearsDf = pd.json_normalize(yearsJsonData['data'])

# 학원 분류 추가
yearsDf['분류'] = yearsDf.apply(lambda x: '학원'
                                if "학원" in x['업종중분류'] else '기타', axis=1)
# 학원 분류일 때 앞에 붙는 '학원-' 삭제
yearsDf['업종중분류'] = yearsDf.apply(lambda x: x['업종중분류'][3:] 
                                     if x['분류']=='학원' and x['업종중분류']!='학원기타' 
                                     else (x['업종중분류'][2:] 
                                     if x['업종중분류']=='학원기타' 
                                     else x['업종중분류']), axis=1)

# 컬럼명/순서 변경
yearsDf.rename({'1년미만':'1년 미만'}, axis=1, inplace=True)
yearsDf.rename({'광역시도':'시도'}, axis=1, inplace=True)
yearsDf = yearsDf.reindex(['시도', '시군구', '업종대분류', '분류', '업종중분류',
                           '1년 미만', '1~2년', '2~3년', '3~5년', '5년 이상'], axis=1)

# 결측치 제거
yearsDf.dropna(inplace=True)

# float to int
yearsDf = yearsDf.astype({'1년 미만': int, '1~2년': int, '2~3년': int, '3~5년': int, '5년 이상': int})

# 합계 컬럼 생성
yearsDf['합계'] = yearsDf['1년 미만'] + yearsDf['1~2년'] + yearsDf['2~3년'] + yearsDf['3~5년'] + yearsDf['5년 이상']

# 평균 컬럼 생성
yearsDf['평균'] = ((0.5 * yearsDf['1년 미만']) + (1.5 * yearsDf['1~2년']) + 
                  (2.5 * yearsDf['2~3년']) + (4 * yearsDf['3~5년']) + (5 * yearsDf['5년 이상'])) / yearsDf['합계']
yearsDf['평균'] = round(yearsDf['평균'], 2)

yearsDf.to_csv("years.csv", index=False, encoding='utf-8-sig')
yearsDf
```
